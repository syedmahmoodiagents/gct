# Graph Management

```python
x = np.linspace(-3, +3, 101)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
yy = np.sin(x)
```

```python
plt.plot(x, y, "r*:")
```

```python
np.random.normal(5,2.1,12)
```

```python
np.random.normal(0,2.1,12)
```

```python
ay = y + np.random.normal(0,0.2,101)
```

```python
plt.plot(x,ay,"r*")
```

```python
plt.plot(x,ay,"r*",ms=20)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))

ax[0, 0].plot(x, y)
ax[0, 0].set_title("Top Left")

ax[0, 1].scatter(x, ay)
ax[0, 1].set_title("Top Right")

ax[1, 0].scatter(x, y)
ax[1, 0].set_title("Bottom Left")

ax[1, 1].plot(x, yy)
ax[1, 1].set_title("Bottom Right")

plt.tight_layout()
# plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
dt = load_iris()
```

```python
df = pd.DataFrame(dt.data, columns=dt.feature_names)
```

```python
df
```

```python
df['sepal length (cm)'].plot(kind='kde')
```

```python
df['sepal length (cm)'].plot.density()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

df['sepal length (cm)'].plot(kind='hist', bins=5, color='purple', ax=ax[0, 0])
ax[0, 0].set_title('Histogram of Sepal Length')

df['sepal width (cm)'].plot(kind='kde', color='purple', ax=ax[0, 1])
ax[0, 1].set_title('Histogram of Sepal Width')

# df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)', color='green', s=100, ax=ax[0, 1])
# ax[0, 1].set_title('')

df['sepal length (cm)'].plot(kind='box', ax=ax[1, 0])
ax[1, 0].set_title('Stock Levels')

df['sepal length (cm)'].plot(kind='bar', color='crimson', ax=ax[1, 1])
ax[1, 1].set_title('Product Returns')

plt.tight_layout()
```

```python
iris = sns.load_dataset("iris")
```

```python

fig, ax = plt.subplots(2, 2, figsize=(10, 10))


sns.histplot(data=iris, x="sepal_length", kde=True, ax=ax[0, 0])
ax[0, 0].set_title("Sepal Length Distribution")


sns.kdeplot(data=iris, x="sepal_width", fill=True, ax=ax[0, 1])
ax[0, 1].set_title("Sepal Width Density")


sns.violinplot(data=iris, x="species", y="petal_length", ax=ax[1, 0])
ax[1, 0].set_title("Petal Length by Species")


sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species", ax=ax[1, 1])
ax[1, 1].set_title("Petal Length vs Width")

plt.tight_layout()
# plt.show()
```
