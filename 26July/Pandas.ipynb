# How to use conditions to create a new column

```python
def find_big_small(x):
    if x > 5:
        print("big")
    else:
        print("small")
```

```python
dnx['DD'].apply(find_big_small)
```

```python
dnx['DD'].apply(lambda x: 'big' if x >= 5 else 'small')
```

```python
dnx
```

```python
dnx['MM'] = dnx['DD'].apply(lambda x: 'big' if x >= 5 else 'small')
```
